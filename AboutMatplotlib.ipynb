# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel('funny numbers')
plt.show()
```

```python
plt.plot([1,3,4,9],[-1,8,1,2],'ro')
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t*2,'r--',t,t**2,'bs')
# plot(x,y,how to show)
```

```python
data = {
    'a' : np.arange(500),
    'b' : np.random.randint(0.50,50),
    'c' : np.random.randn(500)
}
plt.scatter(data['a'],data['c'])
```
